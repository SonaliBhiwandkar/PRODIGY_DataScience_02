```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the Titanic dataset
titanic_data = sns.load_dataset('titanic')
```

```python
# Print the first few rows of the dataset
print(titanic_data.head())
```

```python
# Check for missing values
print(titanic_data.isnull().sum())
```

# Drop the 'deck' column as it has too many missing values

```python
titanic_data = titanic_data.drop('deck', axis=1)
```

# Fill the missing values in the 'age' column with the median age

```python
titanic_data['age'].fillna(titanic_data['age'].median(), inplace=True)
```

# Fill the missing values in the 'embarked' column with the mode

```python
titanic_data['embarked'].fillna(titanic_data['embarked'].mode()[0], inplace=True)
```

# Exploratory Data Analysis

```python
# Visualize the distribution of ages
plt.figure(figsize=(10, 6))
sns.distplot(titanic_data['age'], bins=20, kde=False)
plt.title('Age Distribution')
plt.show()
```

```python
# Visualize the survival rate based on gender
plt.figure(figsize=(8, 6))
sns.barplot(x='sex', y='survived', data=titanic_data)
plt.title('Survival Rate by Gender')
plt.show()
```

```python
# Visualize the survival rate based on passenger class
plt.figure(figsize=(8, 6))
sns.barplot(x='pclass', y='survived', data=titanic_data)
plt.title('Survival Rate by Passenger Class')
plt.show()
```

```python
# Visualize the correlation between numerical features
plt.figure(figsize=(10, 8))
sns.heatmap(titanic_data[['age', 'sibsp', 'parch', 'fare']].corr(), annot=True)
plt.title('Correlation between Numerical Features')
plt.show()
```
